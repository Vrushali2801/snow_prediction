```python
%%bigquery df_task1
SELECT
  station_number,
  wban_number,
  year,
  month,
  day,
  FORMAT_DATE('%Y-%m-%d', DATE(year, month, day)) AS date_ymd,
  mean_temp,
  min_temperature,
  max_temperature,
  total_precipitation,
  snow_depth,
  snow,
  rain,
  fog,
  hail,
  thunder,
  tornado
FROM `bigquery-public-data.samples.gsod`
WHERE year BETWEEN 2000 AND 2005
  AND station_number BETWEEN 725300 AND 726300
```
